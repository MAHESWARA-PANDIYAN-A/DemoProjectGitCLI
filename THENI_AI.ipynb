```python
from sentence_transformers import SentenceTransformer

from sklearn.metrics.pairwise import cosine_similarity
document = '''
Theni is a district located in the southwestern part of Tamil Nadu.
Theni is known for its scenic beauty and agricultural wealth.
The district headquarters is located in Theni town.
Theni is often called the Gateway to Kerala.
Agriculture is the backbone of Theni's economy.
The district is famous for the cultivation of grapes.
Bananas are widely grown in Theni.
Mango cultivation is common in many parts of the district.
Cardamom plantations are found in the hill regions.
The Vaigai River flows through the district.
The Vaigai Dam is a major tourist attraction in Theni.
The dam supports irrigation for nearby farmlands.
Cumbum Valley is one of the most fertile regions in Tamil Nadu.
Cumbum is famous for grape production.
Uthamapalayam is an important town in the district.
Bodinayakanur is known as the Cardamom City of India.
Bodinayakanur is surrounded by the Western Ghats.
The Western Ghats add natural beauty to Theni district.
Meghamalai is a popular hill station in Theni.
Meghamalai is known for its tea estates and forests.
Suruli Falls is a famous waterfall in the district.
Suruli Falls attracts tourists throughout the year.
Theni has a pleasant climate compared to many other districts.
The district shares a border with the state of Kerala.
Tourism plays an important role in Theni's economy.
The district is rich in biodiversity and wildlife.
Several temples in Theni attract devotees from different regions.
Theni is well connected by road to major cities in Tamil Nadu.
The people of Theni are known for their hospitality and culture.
Theni continues to grow while preserving its natural beauty and traditions.
'''
#print(document)

chunks = document.split(".")
#print(chunks)

chunks = [chunk.strip() for chunk in chunks if chunk.strip()]

#print(chunks)

model = SentenceTransformer('all-MiniLM-L6-v2')
sentence = "Agriculture is the backbone of Theni's economy."
embedding = model.encode(sentence)
#print(embedding.shape)

# Generate embeddings for all chunks
embeddings = model.encode(chunks)
#print(embeddings.shape)

documents = [
    "Theni is a district located in the southwestern part of Tamil Nadu",
    "Theni is known for its scenic beauty and agricultural wealth",
    "The district headquarters is located in Theni town",
    "Theni is often called the Gateway to Kerala",
    "Agriculture is the backbone of Theni's economy",
    "The district is famous for grape cultivation",
    "Bananas are widely grown in Theni district",
    "Mango cultivation is common in many parts of Theni",
    "Cardamom plantations are found in the hill regions",
    "The Vaigai River flows through the district",
    "The Vaigai Dam is a major tourist attraction",
    "The dam supports irrigation for nearby farmlands",
    "Cumbum Valley is one of the most fertile regions in Tamil Nadu",
    "Cumbum is famous for grape production",
    "Uthamapalayam is an important town in the district",
    "Bodinayakanur is known as the Cardamom City of India",
    "Bodinayakanur is surrounded by the Western Ghats",
    "The Western Ghats add natural beauty to Theni district",
    "Meghamalai is a popular hill station in Theni",
    "Meghamalai is known for its tea estates and forests",
    "Suruli Falls is a famous waterfall in the district",
    "Suruli Falls attracts tourists throughout the year",
    "Theni has a pleasant climate compared to many other districts",
    "The district shares a border with Kerala",
    "Tourism plays an important role in Theni's economy",
    "The district is rich in biodiversity and wildlife",
    "Several temples in Theni attract devotees from different regions",
    "Theni is well connected by road to major cities",
    "The people of Theni are known for their hospitality and culture",
    "Theni continues to preserve its natural beauty and traditions",
    "Tea plantations are common in the hill areas",
    "Coffee is also cultivated in some regions of Theni",
    "The district receives visitors from across India",
    "Agriculture remains the primary occupation of many residents",
    "Theni is famous for its fresh fruits and vegetables",
    "The region has several scenic viewpoints",
    "Forests in Theni support diverse plant and animal species",
    "The district contributes significantly to Tamil Nadu agriculture",
    "Traditional festivals are celebrated with enthusiasm in Theni",
    "Theni is one of the beautiful districts of Tamil Nadu"
]

query = input('Enter Query : ')
#print(query)

query_embedding = model.encode(query)
#print(query_embedding.shape)
#print(query_embedding)

similarities=cosine_similarity([query_embedding],embeddings)
#print(similarities)

best_match_index=similarities.argmax()
#print(chunks[best_match_index])

retrived_chunks = chunks[best_match_index]

prompt = f'''
Question : {query}?
Context : {retrived_chunks}
'''

print(prompt)
```
