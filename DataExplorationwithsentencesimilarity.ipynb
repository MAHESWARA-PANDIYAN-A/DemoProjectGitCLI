```python
import csv
import random

f = r"E:\AgenticAIInternship\data\movie-dialog-corpus\movie_lines.tsv"

texts = []
with open(f, "r", encoding="utf-8") as i:
    csv_reader = csv.reader(i, delimiter="\t")
    for line in csv_reader:
        texts.append(line[-1])
random.seed(1)
random.shuffle(texts)
texts = texts[:10000]
print(texts[:10])
```
